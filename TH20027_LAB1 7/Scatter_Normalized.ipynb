```python
min_max_scaler = preprocessing.MinMaxScaler()
normalized = min_max_scaler.fit_transform(df)
print(normalized)
```

```python
normalized_df = pd.DataFrame(normalized, columns=df.columns)
print(normalized_df)
```

```python
df = normalized_df.iloc[:, 0:10]
Target = normalized_df.Type
df,Target
```
